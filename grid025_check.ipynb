# Check the CICE grid coordinates

```python
import xarray as xr, numpy as np
import matplotlib.pyplot as plt
```

## 1°

```python
om2cice1_grid = xr.open_dataset('/g/data/ik11/inputs/access-om2/input_20201102/cice_1deg/grid.nc')
om2mom1_grid = xr.open_dataset('/g/data/ik11/inputs/access-om2/input_20201102/mom_1deg/ocean_hgrid.nc')
```

```python
# Comparison with MOM supergrid
print("Coordinate errors for 1 degree grid")
diff = np.rad2deg(om2cice1_grid.tlat.values) - om2mom1_grid.y[1::2,1::2]
print('TLAT % 10.3e' % abs(diff).max().values)
diff = np.rad2deg(om2cice1_grid.tlon.values) - om2mom1_grid.x[1::2,1::2]
print('TLON % 10.3e' % abs(diff).max().values)

# NE points for U
diff = np.rad2deg(om2cice1_grid.ulat.values) - om2mom1_grid.y[2::2,2::2]
print('ULAT % 10.3e' % abs(diff).max().values)
diff = np.rad2deg(om2cice1_grid.ulon.values) - om2mom1_grid.x[2::2,2::2]
print('ULON % 10.3e' % abs(diff).max().values)
```

```python
fig, axes = plt.subplots()
diff = np.rad2deg(om2cice1_grid.ulat.values) - om2mom1_grid.y[2::2,2::2]
diff.plot()
axes.set_title('1° grid ULAT error')
```

```python
fig, axes = plt.subplots()
diff = np.rad2deg(om2cice1_grid.tlat.values) - om2mom1_grid.y[1::2,1::2]
diff.plot()
axes.set_title('1° grid TLAT error')
```

```python
fig, axes = plt.subplots()
diff = np.rad2deg(om2cice1_grid.ulon.values) - om2mom1_grid.x[2::2,2::2]
diff.plot()
axes.set_title('1° grid ULON error')
```

```python
fig, axes = plt.subplots()
diff = np.rad2deg(om2cice1_grid.tlon.values) - om2mom1_grid.x[1::2,1::2]
diff.plot()
axes.set_title('1° grid TLON error')
```

```python
fig, axes = plt.subplots()
diff = (np.rad2deg(om2cice1_grid.ulat.values) - om2mom1_grid.y[2::2,2::2])/om2mom1_grid.y[2::2,2::2]
diff.plot()
axes.set_title('1° grid ULAT relative error')
```

```python
fig, axes = plt.subplots()
diff = (np.rad2deg(om2cice1_grid.tlat.values) - om2mom1_grid.y[1::2,1::2])/om2mom1_grid.y[1::2,1::2]
diff.plot()
axes.set_title('1° grid TLAT relative error')
```

```python
fig, axes = plt.subplots()
diff = (np.rad2deg(om2cice1_grid.ulon.values) - om2mom1_grid.x[2::2,2::2])/om2mom1_grid.x[2::2,2::2]
diff.plot()
axes.set_title('1° grid ULON relative error')
```

```python
fig, axes = plt.subplots()
diff = (np.rad2deg(om2cice1_grid.tlon.values) - om2mom1_grid.x[1::2,1::2])/om2mom1_grid.x[1::2,1::2]
diff.plot()
axes.set_title('1° grid TLON relative error')
```

## 0.25°

```python
om2cice025_grid = xr.open_dataset('/g/data/ik11/inputs/access-om2/input_20230515_025deg_topog/cice_025deg/grid.nc')
om2mom025_grid = xr.open_dataset('/g/data/ik11/inputs/access-om2/input_20230515_025deg_topog/mom_025deg/ocean_hgrid.nc')
```

```python
# Comparison with MOM supergrid
print("Coordinate errors for 0.25 degree grid")
diff = np.rad2deg(om2cice025_grid.tlat.values) - om2mom025_grid.y[1::2,1::2]
print('TLAT % 10.3e' % abs(diff).max().values)
diff = np.rad2deg(om2cice025_grid.tlon.values) - om2mom025_grid.x[1::2,1::2]
print('TLON % 10.3e' % abs(diff).max().values)

# NE points for U
diff = np.rad2deg(om2cice025_grid.ulat.values) - om2mom025_grid.y[2::2,2::2]
print('ULAT % 10.3e' % abs(diff).max().values)
diff = np.rad2deg(om2cice025_grid.ulon.values) - om2mom025_grid.x[2::2,2::2]
print('ULON % 10.3e' % abs(diff).max().values)
```

```python
# NW points for U
print("Coordinate differences for 0.25 grid using NW corner")
diff = np.rad2deg(om2cice025_grid.ulat.values) - om2mom025_grid.y[2::2,:-1:2]
print('ULAT % 10.3e' % abs(diff).max().values)
diff = np.rad2deg(om2cice025_grid.ulon.values) - om2mom025_grid.x[2::2,:-1:2]
print('ULON % 10.3e' % abs(diff).max().values)
```

```python
fig, axes = plt.subplots()
diff = np.rad2deg(om2cice025_grid.ulat.values) - om2mom025_grid.y[2::2,2::2]
diff.plot()
axes.set_title('0.25 grid ULAT error')
```

```python
# number of points with each difference in ULAT
unique, counts = np.unique(np.rad2deg(om2cice025_grid.ulat.values) - om2mom025_grid.y[2::2,2::2], return_counts=True)
dict(zip(unique, counts))
```
